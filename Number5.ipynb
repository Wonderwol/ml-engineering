```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
inf = pd.read_csv("S:\ФБИ-11\ml-engineering\data.csv") # "чтение" файла
```

```python
inf["нет ответа"]=inf["Ответ"].isna()
```

```python
inf = inf.groupby(['Тип отзыва', 'Подведомственность'])['нет ответа'].sum().reset_index()
sns.barplot(x='Тип отзыва', y='нет ответа', hue='Подведомственность', data=inf)
```
